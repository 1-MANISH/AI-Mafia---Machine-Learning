```python
import numpy as np

x=np.arange(-10,10,0.01)#gives point b/w -10 to 10 with gap 0.01
# print(len(x),x)
y1=np.sin(x)
y2=x*x+5*x+2
y3=x**3
y4=np.cos(x)

plt.subplot(2,2,1)
plt.plot(x,y1,'r')

plt.subplot(2,2,2)
plt.plot(x,y3,'blue')

plt.subplot(2,2,3)
plt.plot(x,y2,'r--')

plt.subplot(2,2,4)
plt.plot(x,y4,'r-.')

plt.show()
```

# Bar plot/chart

```python
'''
--A bar is very commonly used 2 d data visualization made of rectugangler
bars
--
    plt.plot(x,y,width,color,label)
    
'''

```

```python
x=[i for i in range(10)]
y=[4,3,7,2,9,0,12,5,6,8]

plt.bar(x,y)
plt.show()
```

```python
x=[i for i in range(10)]
y=[4,3,7,2,9,0,12,5,6,8]

plt.bar(x,y,color="green",width=0.3)
plt.show()
```

```python
x=[i for i in range(10)]
y=[4,3,7,2,9,0,12,5,6,8]
z=[2,7,3,1,12,3,10,4,3,15]
plt.bar(x,y,color="green",width=0.3)
plt.bar(x,z,color="blue",width=0.3)
plt.show()
```

```python
x1=[i for i in range(10)]
y=[4,3,7,2,9,0,12,5,6,8]

x2=[i+0.3 for i in range(10)]
z=[2,7,3,1,12,3,10,4,3,15]

plt.bar(x1,y,color="red",width=0.3)
plt.bar(x2,z,color="blue",width=0.3)
plt.show()
```

```python
Rahul_score=[56,0,13,45,56]
Rohit_score=[90,56,3,34,67]

x1=[i for i in range(1,6)]
x2=[i+0.4 for i in range(1,6)]

plt.xlabel("Matches")
plt.ylabel("Scores")
plt.bar(x1,Rahul_score,color="Blue",width=0.4,label="Rahul-score")
plt.bar(x2,Rohit_score,color="red",width=0.4,label="Rohit-score")
# To label
plt.legend()
plt.show()
```

# Pie chart

```python
'''
A pie plot is circuler plot ,divided into slices
to show numerical proportion 
Pie plot widley used in the buisenss world


-- plt.pie(sizes   ,  color,  labels,  shadow,  explode )
            array     array    array                 bahr kerna
 
'''

```

```python
hours=[2,6,1,4,2,6,2,1]
activity=["Ready","School","Rest","Study","Play","Sleep","Tv","Woke"]
plt.pie(hours,labels=activity)
plt.show()
```

```python
hours=[2,6,1,4,2,6,2,1]
activity=["Ready","School","Rest","Study","Play","Sleep","Tv","Woke"]
plt.pie(hours,labels=activity,shadow=True)
plt.show()
```

```python
hours=[2,6,1,4,2,6,2,1]
activity=["Ready","School","Rest","Study","Play","Sleep","Tv","Woke"]
explode=[0.08,0,0,0,0,0.04,0,0.08]
plt.pie(hours,labels=activity,shadow=True,explode=explode)
plt.show()
```
